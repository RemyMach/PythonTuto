```python
import csv

f = open("guns.csv")
csvreader = csv.reader(f)
guns = list(csvreader)
print(guns[0:5])
header = guns[:1]
guns = guns[1:]
```

## 2. Compter le nombre de dcs par armes  feu aux USA par anne

```python
years = [ gun[1] for gun in guns]
print(years[:5])
```

```python
year_counts = {}
for year in years:
    if year in year_counts:
        year_counts[year] += 1
    else:
        year_counts[year] = 1
print(year_counts)
```

```python
#on gagne une ligne
year_counts1 = {}
for year in years:
    if year not in year_counts1:
        year_counts1[year] = 0
    year_counts1[year] += 1
print(year_counts1)
```

```python
import datetime

dates = [datetime.datetime(year = int(row[1]), month = int(row[2]), day = 1) for row in guns]
print(dates[:5])
```

```python
date_counts = {}
for date in dates:
    if date not in date_counts:
        date_counts[date] = 0
    date_counts[date] += 1
print(date_counts)
```

```python
sex_counts = {}
race_counts = {}

for gun in guns:
    if gun[7] is not in race_counts:
        sex_counts[gun[7]] = 0
    sex_counts[gun[7]] += 1
    if gun[5] is not in sex_counts:
        sex_counts[gun[5]] = 0
    sex_counts[gun[5]] += 1
    
```
