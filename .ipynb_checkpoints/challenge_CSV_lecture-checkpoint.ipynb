# Challenge

Ouvrir un fichier sur jupyter et créer une liste d'entier contenant le nombre d'habitants de tous les départements français

```python
file_csv = open("departement.csv","r",encoding='ISO-8859-1')
result = file_csv.read()
print(result)

list = []

for row in result:
    split_row = row.split(',')
    list.append(split_row)

print(list)
```
