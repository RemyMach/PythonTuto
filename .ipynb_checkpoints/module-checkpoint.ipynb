## Compter les victoires pour une année donnée

```python
def nfl_wins(team, year):
    dictionnaire_nfl = {}
    for cellule in nfl:
        team_not_choose = cellule[2]
        year_not_choose = cellule[0]
        if team == team_not_choose and int(year_not_choose) == year:
            if team in dictionnaire_nfl:
                dictionnaire_nfl[team] += 1
            else:
                dictionnaire_nfl[team] = 1
    return dictionnaire_nfl

print(nfl_wins('Minnesota Vikings', 2009))
```
