# Python et les dates

## Le module Time

```python
# time.time()
import time
current_time = time.time()
#timestamp
print(current_time)
```

```python
# time.gmtime()
current_struct_time = time.gmtime(current_time)
print(current_struct_time)
```

```python
current_hour = current_struct_time.tm_hour
print(current_hour)
```

```python
current_year = current_struct_time.tm_year
print(current_year)
```

## le module datetime

```python
# datetime.now()

import datetime

current_datetime = datetime.datetime.now()
current_year = current_datetime.year
current_month = current_datetime.month

print(current_datetime)
print(current_year)
print(current_month)
```

# fonction timedelta

```python
today = datetime.datetime.now()
diff = datetime.timedelta(weeks = 1, days = 5)
print(diff)
```

```python
result = today + diff
print(result)
```

```python
diff = datetime.timedelta(days=1)
tomorrow = today + diff
yesterday = today - diff
```

```python
print(tomorrow)
print(yesterday)
```

## Exercice

```python
import csv
f = open("askreddit.csv")
csvreader = csv.reader(f)
posts = list(csvreader)
posts = posts[1:]
print(posts[0:5])
```

```python
# convertir en objet datetime
# datetime.fromtimestamp()

datetime_object = datetime.datetime.fromtimestamp(1440082910.0)
print(datetime_object)
```

```python
for post in posts:
    Time = post[2]
    try:
        Time = float(Time)
        post[2] = datetime.datetime.fromtimestamp(Time)
    except:
        Time = float(Time)
        post[2] = datetime.datetime.fromtimestamp(Time)

print(posts[0:5])
```

## Compter les posts de Mai

```python
may_count = 0
for post in posts:
    if post[2].month == 5:
        may_count += 1
print(may_count)
```

## Compter les posts pour nimporte quel mois

```python
def count_post_month(month):
    month_count = 0
    for post in posts:
        if post[2].month == month:
            month_count += 1
    return month_count

for x in range(12):
    print("Mois: %s, résultat: %s" % (x+1, count_post_month(x+1)))
```
