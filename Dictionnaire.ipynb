## compter le nombre d'itération dans une liste et présenter dans un dictionnaire

```python
PLANET_NAMES = ["Mercure", "venus", "Terre", "Mars", "Jupiter", "Saturne"
                , "Neptune", "venus", "Uranus", "venus"]
planet_items = {}

for item in PLANET_NAMES:
    if item in planet_items:
        planet_items[item] += 1
    else:
        planet_items[item] = 1
print(planet_items)
```

## lire un fichier texte avec un encodage spécial

```python
vocabulary = open('dictionnaire.txt','r', encoding="utf-8").read()
print(vocabulary)
```

```python
tokenized_vocabulary = vocabulary.split(' ')
print(tokenized_vocabulary)
tokenized_vocabulary_nospace = []
for word in tokenized_vocabulary:
    if len(word) != 0 and word != " ":
        tokenized_vocabulary_nospace.append(word)
print(tokenized_vocabulary_nospace[0:5])
```
