# Gestion des erreurs

## extraire des éléments unique avec la fonction set

### set

```python
animals = ['chien', 'tigre', 'chien', 'chat', 'tigre']
unique_animals = set(animals)
print(unique_animals)
```

### avec set la fonction devient un objet sert ce n'est ni une liste ni un dictionnaire

```python
# add()
unique_animals.add("Tortue")
print(unique_animals)
```

```python
# remove()
unique_animals.remove('tigre')
print(unique_animals)
```

```python
# convertir un objet set en list 
# list()
unique_animals = list(unique_animals)
print(unique_animals)
```

```python
import csv
f = open('legislators.csv')
csvreader = csv.reader(f)
legislators = list(csvreader)
legislators = legislators[1:]
gender = []
for legislator in legislators:
    gender.append(legislator[3])
gender = set(gender)
print(gender)
```

# Quand python lève une erreur il instencie automatiquement un objet de la classe Exception donc on peut disposer de ses méthodes

```python
try:
    int('')
except Exception:
    print('Impossible de convertir')
```

```python
try:
    int('')
except Exception as exc:
    print(type(exc))
    print(str(exc))
```

```python
int_years = []

for year in birth_years:
    try:
        year = int(year)
    except Exception as exc:
        pass
    int_years.append(year)
print(int_years)
```
