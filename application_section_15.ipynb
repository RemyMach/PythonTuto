```python
import csv

f = open('nfl_suspensions_data.csv')
csvreader = csv.reader(f)
nfl_suspensions = list(csvreader)
nfl_suspensions = nfl_suspensions[1:]
year = {}
count_year = None
for nfl_suspension in nfl_suspensions:
    row_year = nfl_suspension[5]
    if row_year is not None and len(row_year) != 0:
        if row_year in year:
            year[row_year] += 1
        else:
            year[row_year] = 1

print(year)
```
