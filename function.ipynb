# function

```python
def clean_text(text):
    text = text.replace('\n','')
    text = text.replace(',','')
    text = text.replace('.','')
    text = text.replace("'","")

    return text
```

## Arguments multiples

```python
def modify_text(text_string, modify_string, replacement_string):
    cleaned_string = text_string.replace(modify_string, replacement_string)
    return cleaned_string

goal = modify_text('GOOOOOAAAAAALLLLL.................','.','!')
print(goal)
```

```python
def clean_text_modify(text, list_caracters, replacement_string):
    for caracter in list_caracters:
        text = text.replace(caracter, replacement_string)

    return text

text = "je,ne,suis.pas.toi \n sombre idiot"
caracters = [".",",","\n"]
replacement_caracter = " "

texte_result = clean_text_modify(text, caracters, replacement_caracter)
print(texte_result)
```

## ARGUMENT AVEC UNE VALEUR PAR DFAUT

```python
vocabulary_text_txt = open('texte.txt','r', encoding="utf-8").read()
#print(vocabulary_text_txt)

def tokenize(string, special_characters, replacement_string, clean = False):
    cleaned_text = string
    if clean:
        print('pomme')
        cleaned_text = clean_text_modify(string, special_characters, replacement_string)
    text_tokens = cleaned_text.split(' ')
    return text_tokens

clean_characters = [".",",","\n"]
replacement = ""
tokenized_text = tokenize(vocabulary_text_txt, clean_characters, replacement, True)
print(tokenized_text[0:10])
```
