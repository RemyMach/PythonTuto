## Ecrire une boucle qui permet de vérifier la présence des éléments de weather_types in new_weather

```python
weather_types = ['Pluie','Soleil','Nuage','Nuage-Pluie','Orage','Climat']
weather_type_found = []
for weather in new_weather:
    list_weather = weather.split(',')
    if list_weather[1] in weather_types:
        weather_type_found.append(weather)
print(weather_type_found[0:10])
```

```python
PLANET_NAMES = ["Mercure", "venus", "Terre", "Mars", "Jupiter", "Saturne"
                , "Neptune", "Uranus"]
long_names = []
short_names = []

for planet in PLANET_NAMES:
    if len(planet) > 5:
        long_names.append(planet)
    else:
        short_names.append(planet)
print(long_names)
print(short_names)
```
