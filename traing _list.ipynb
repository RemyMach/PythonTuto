# Projet: Explorer les naissances aux US depuis 2000

```python
def read_csv(file):
    us_birth_string = open(file,'r', encoding="utf-8").read()
    us_birth_string = us_birth_string.split('\n')
    #on enlève l'en-tête
    us_birth_string = us_birth_string[1:]
    
    final_list = []
    
    for cellule in us_birth_string:
        int_fields = []
        string_list = cellule.split(',')
        for row in string_list:
            int_fields.append(int(row))
        final_list.append(int_fields)
    return final_list

us_birth_list = read_csv('US-birth.csv')
print(us_birth_list[0:10])
```

```python
def month_births(us_birth_list):
    us_month_temporary = {}
    for row in us_birth_list:
        month = row[1]
        birth = row[4]
        if month in us_month_temporary:
            us_month_temporary[month] += birth
        else:
            us_month_temporary[month] = birth
    return us_month_temporary

us_month_birth = month_births(us_birth_list)
print(us_mont_birth)
```

```python
def day_births(us_birth_list):
    us_day_temporary = {}
    for row in us_birth_list:
        day = row[3]
        birth = row[4]
        if day in us_day_temporary:
            us_day_temporary[day] += birth
        else:
            us_day_temporary[day] = birth
    return us_day_temporary

us_day_birth = day_births(us_birth_list)
print(us_day_birth)
```

```python
def calc_counts(us_birth_list, indice):
    list_possible_value = ['year', 'month', 'day_of_month', 'day']
    if indice not in list_possible_value:
        return
    value_indice_convert = list_possible_value.index(indice)
    us_day_temporary = {}
    for row in us_birth_list:
        indice_value = row[value_indice_convert]
        birth = row[4]
        if indice_value in us_day_temporary:
            us_day_temporary[indice_value] += birth
        else:
            us_day_temporary[indice_value] = birth
    return us_day_temporary

us_day_birth = calc_counts(us_birth_list, 'day')
print(us_day_birth)
```
